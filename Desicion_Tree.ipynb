```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/temps.csv")
dataset
```

```python
dataset.head(5)
```

```python
dataset.shape
```

```python
#one hot encoding
dataset = pd.get_dummies(dataset)
dataset
```

```python
dataset.shape
```

```python
#separate feature and labels
labels = np.array(dataset['actual'])
features = dataset.drop('actual',axis=1)
features =np.array(features)
features.shape
```

```python
from sklearn.model_selection import train_test_split
train_features,test_feature,train_labels,test_labels = train_test_split(features,labels,test_size =0.20)
```

```python
#train the model
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=1000)
```

```python
rf.fit(train_features,train_labels)
```

```python
prediction = rf.predict(test_feature)
prediction
```

```python
errors = abs(prediction-test_labels)
errors
```

```python
mae = (errors/test_labels)
mae
```

```python
accuracy = (1-np.mean(mae))*100
accuracy
```

```python
from sklearn.tree import export_graphviz
tree = rf.estimators_[50]
export_graphviz(tree,out_file='tree.dot')
```
