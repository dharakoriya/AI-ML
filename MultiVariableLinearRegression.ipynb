```python
#definition of normalization function
def normalization(X):
  mean=np.mean(X,axis=0) #axis=0 means row, axis=1 means column; if you don't mention axis, it will calculate mean of whole dataset
  std=np.std(X,axis=0)
  XNorm=(X-mean)/std
  return XNorm,mean,std
```

```python
#normalizing the dataset
XArray=dataset.values #for converting dataset from dataframe to array
size=len(XArray[:,-1]) #why -1? all rows, but last column to count no of rows
x2=XArray[:,0:2].reshape(size,2)
XInput,mean,std=normalization(x2)
y=XArray[:,-1].reshape(size,1)
```

```python
def hypothesis(thetaList,x1,x2):
  Hx=thetaList[2]*x2+thetaList[1]*x1+thetaList[0]
  return Hx
```

```python
def costfunction(thetaList,x1,x2,y,m):
  totalCost=0
  for i in range(m):
    totalCost+=(thetaList[2]*x2[i]+thetaList[0]+thetaList[1]*x1[i]-y[i])**2
  return (totalCost/(2*m))
```

```python
def gradientdescent(thetaList,x1,x2,y,alpha,m):
  summationTheta0=0
  summationTheta1=0
  summationTheta2=0

  for i in range(m):
    summationTheta0+=(thetaList[2]*x2[i]+thetaList[1]*x1[i]+thetaList[0])-y[i]
    summationTheta1+=x1[i]*((thetaList[2]*x2[i]+thetaList[1]*x1[i]+thetaList[0])-y[i])
    summationTheta2+=x2[i]*((thetaList[2]*x2[i]+thetaList[1]*x1[i]+thetaList[0])-y[i])


  updatedTheta0=thetaList[0]-(alpha/m)*summationTheta0
  updatedTheta1=thetaList[1]-(alpha/m)*summationTheta1
  updatedTheta2=thetaList[2]-(alpha/m)*summationTheta2

  updatedThetaList=[updatedTheta0,updatedTheta1,updatedTheta2]

  print(updatedThetaList)
  return updatedThetaList
```

```python
def training(x1,x2,y,alpha,epochs):
  theta0=0
  theta1=0
  theta2=0

  thetaList=[theta0,theta1,theta2]
  m=x1.size

  costfunctionValues=[]
  for i in range(epochs):
    thetaList=gradientdescent(thetaList,x1,x2,y,alpha,m)
    costfunctionValues.append(costfunction(thetaList,x1,x2,y,m))
    print("epochs: ",i,thetaList,costfunctionValues[-1])

  return costfunctionValues,thetaList
```

```python
alpha=0.01
epochs=500
costfunctionValues,thetaList=training(XInput[:,0:1],XInput[:,0:2],y,alpha,epochs)
```
