Design linear regression model for univariant

```python
def hypothesis(thetaList,x):
  Hx=thetaList[1]*x+thetaList[0]
  return Hx
```

```python
def costfunction(thetaList,x,y,m):
  totalCost=0
  for i in range(m):
    totalCost+=(thetaList[0]+thetaList[1]*x[i]-y[i])**2
  return (totalCost/(2*m))
```

```python
def gradientdescent(thetaList,x,y,alpha,m):
  summationTheta0=0
  summationTheta1=0
  for i in range(m):
    summationTheta0+=(thetaList[1]*x[i]+thetaList[0])-y[i]
    summationTheta1+=x[i]*((thetaList[1]*x[i]+thetaList[0])-y[i])

  updatedTheta0=thetaList[0]-(alpha/m)*summationTheta0
  updatedTheta1=thetaList[1]-(alpha/m)*summationTheta1

  updatedThetaList=[updatedTheta0,updatedTheta1]

  print(updatedThetaList)
  return updatedThetaList
```

```python
def training(x,y,alpha,epochs):
  theta0=0
  theta1=0
  thetaList=[theta0,theta1]
  m=x.size

  costfunctionValues=[]
  for i in range(epochs):
    thetaList=gradientdescent(thetaList,x,y,alpha,m)
    costfunctionValues.append(costfunction(thetaList,x,y,m))
    #plotting the linear line over the scattered datapoints
    yPred=thetaList[0]+thetaList[1]*x
    plt.plot(x,yPred,'r')
    plt.scatter(x,y)
    plt.show()
  return costfunctionValues
```

```python
trainingData=pd.read_csv("/content/train.csv")
cleanedData=trainingData.dropna()
cleanedData=cleanedData.values #to convert to
xFeature=cleanedData[:,0]
yFeature=cleanedData[:,1]
alpha=0.0005
epochs=50
costfunctionValues=training(xFeature,yFeature,alpha,epochs)

x=np.arange(0,len(costfunctionValues),step=1)
plt.plot(x,costfunctionValues)
plt.title("Learning Curve")
plt.xlabel("Number Of Iterations")
plt.ylabel("Cost Function Value")
plt.show
```
